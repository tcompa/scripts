```python
%matplotlib inline
import matplotlib
import numpy
import random
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.figsize'] = [12.0, 5.0]
rcParams['figure.dpi'] = 256
```

```python
for i in range(10):
    data = numpy.random.normal(random.uniform(0.0, 1.0), random.uniform(0.2, 0.5), size=100000)
    h, b = numpy.histogram(data, bins=100, normed=True)
    b = 0.5 * (b[1:] + b[:-1])
    plt.plot(b, h, label='data %i' % i)
plt.legend()
plt.show()
```
